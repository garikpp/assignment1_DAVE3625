```python
%matplotlib inline
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
import re

url = 'https://raw.githubusercontent.com/umaimehm/Intro_to_AI_2021/main/assignment1/Ruter_data.csv' #link til ruter datasett

df = pd.read_csv(url, sep=';')


#plot av antall bussruter i hver kommune
plt.title("Antall bussruter i hver kommune")
df["Kommune"].value_counts()
sns.countplot(x='Kommune', data=df);
plt.xticks(rotation=90);
plt.show()

plt.title("Antall bussruter i hver fylke")
df["Fylke"].value_counts()
sns.countplot(x='Fylke', data=df);
plt.xticks(rotation=90);
plt.show()

plt.title("Busstyper")
df["Linjetype"].value_counts()
sns.countplot(x='Linjetype', data=df);
plt.xticks(rotation=90);
plt.show()

plt.title("Buss Kapasitet")
df["Kjrety_Kapasitet"].plot.box()
plt.show()

plt.title("Antall Passasjerere ombord")
plt.hist(df.Passasjerer_Ombord, bins=7, rwidth=0.8)
plt.xlabel('Passasjerer Ombord')
plt.ylabel('Total')
plt.show()

plt.title("Buss Kapasitet")
plt.hist(df.Kjrety_Kapasitet, bins=7, rwidth=0.7)
plt.xlabel('Buss Kapasitet')
plt.ylabel('Total')
plt.show()
```
